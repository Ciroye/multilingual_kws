```python
WORD = "go"
GC = f"./eleven_word_dataset/extractions_gcloud/{WORD}/*.wav"
DS = f"./eleven_word_dataset/extractions_deepspeech/{WORD}/*.wav"

gc = glob.glob(GC)
ds = glob.glob(DS)

gc_bp = [os.path.basename(f) for f in gc]
ds_bp = [os.path.basename(f) for f in ds]
print("lengths:",len(gc), len(ds))
intersection = len(set(gc_bp).intersection(ds_bp))
print("intersection:", intersection)
dif = set(gc_bp).difference(ds_bp)
print("difference:", len(dif))
dif = list(dif)
```

```python
words = ["up","down","left","right","stop","go","off","on","yes","no"]
len(words)
intersections = []
for w in words:
    GC = f"./eleven_word_dataset/extractions_gcloud/{w}/*.wav"
    DS = f"./eleven_word_dataset/extractions_deepspeech/{w}/*.wav"

    gc = glob.glob(GC)
    ds = glob.glob(DS)
    
    gc_bp = [os.path.basename(f) for f in gc]
    ds_bp = [os.path.basename(f) for f in ds]
    #print("lengths:",len(gc), len(ds))
    gc_bp = set(gc_bp)
    ds_bp = set(ds_bp)
    intersection = gc_bp.intersection(ds_bp)
    difference = gc_bp.difference(ds_bp)
    intersections.append(len(intersection))
    #print("deepspeech vs intersections", len(ds) - intersection)
    results = [w,len(gc_bp), len(ds_bp),len(intersection), len(difference)]
    print(",".join([str(r) for r in results]))

print("intersections")
for x in list(zip(words, intersections)):
    print(x)
```
